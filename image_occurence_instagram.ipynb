```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
```

```python
input_path = './data/csv/haneul_instagram_labels.csv'
```

```python
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['font.family'] = 'appleGothic'
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
instagram_df = pd.read_csv(input_path)
```

```python
instagram_df
```

```python
instagram_top_labels1 = instagram_df.head(10)
```

```python
instagram_top_labels1
```

```python
plt.figure(figsize=(12, 9))
plt.bar(instagram_top_labels1['label'], instagram_top_labels1['frequency'])
plt.xlabel('Label', loc='right')
plt.ylabel('Frequency', loc='top')
plt.title('Instagram Label Occurrence')
plt.xticks(rotation=45, ha="right")
plt.savefig('./result/instagram_label_occurrence.png', bbox_inches='tight')
plt.show()
```

```python
from wordcloud import WordCloud
```

```python
instagram_top_labels2 = instagram_df.head(50)
```

```python
wordcloud = WordCloud(width=1000, height=1000, background_color='white').generate_from_frequencies(instagram_top_labels2.set_index('label')['frequency'].to_dict())
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.savefig('./result/instagram_label_wordcloud.png')
plt.show()
```
